```python
#Obtain iris dataset
#The Iris Data set is a dataset of three types of Iris plants. 
#Number of Input Patterns: 150 (50 in each of three classes)
#number of Attributes or Input Features: 4 numeric, predictive attributes and the class
from sklearn import datasets    #module for ml, data analytics etc
iris = datasets.load_iris()     
X = iris.data
y = iris.target
print(X.shape)
print(y.shape)
```

```python
# Z-normalized using the StandardScaler object from
# the module “sklearn.preprocessing.”
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = (scaler.transform(X))
print(X.shape)
```

```python
#Visualizing the Iris dataset:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Convert target labels to categorical values
species_names = ['Setosa', 'Versicolor', 'Virginica']
y_labels = [species_names[i] for i in y]
# Create a Pandas DataFrame for easier data manipulation
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = y_labels
# Visualize the dataset using a pair plot
sns.pairplot(df, hue='species')
# Visualize the distribution of each feature using histograms
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.hist(df['sepal length (cm)'], bins=20)
plt.title('Sepal Length')

plt.subplot(2, 2, 2)
plt.hist(df['sepal width (cm)'], bins=20)
plt.title('Sepal Width')

plt.subplot(2, 2, 3)
plt.hist(df['petal length (cm)'], bins=20)
plt.title('Petal Length')

plt.subplot(2, 2, 4)
plt.hist(df['petal width (cm)'], bins=20)
plt.title('Petal Width')
plt.tight_layout()
plt.show()
```

```python
#Split into training, validation, and test set:

from sklearn.model_selection import train_test_split
X_temp, X_test, y_temp, y_test = train_test_split(\
    X,y,test_size = 20, random_state = 123, stratify = y)  
# sets aside 20 samples for testing. The remaining data, X_temp and y_temp, 
#which is now 130 is split into 20 for validation (X_valid, y_valid) and the rest 110 for training 
#(X_train, y_train).

X_train, X_valid, y_train, y_valid = train_test_split(\
    X_temp,y_temp,test_size = 20, random_state = 123, stratify = y_temp)  # pass  

print(X_train.shape)
print(y_train.shape)
# optional to free up some memory by deleting non-used arrays:
del X_temp,y_temp, X,y
```
